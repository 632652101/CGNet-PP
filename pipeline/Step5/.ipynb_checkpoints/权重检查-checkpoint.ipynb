# 源网络权重获取

```python
import torch
import pickle
```

```python
# 作者模型保存的时候是按cuda的方式进行存储的, 加载的时候需要按 CPU 的方式加载
state_dict = torch.load("../weights/model_cityscapes_train_on_trainset.pth", map_location=torch.device('cpu'))
state_dict
```

```python
weight_dict = state_dict["model"]
npy = []
for k, v in weight_dict.items():
    if v.cpu().numpy().shape != ():
        npy.append(v.cpu().numpy())
with open("weight.pkl", "wb") as fp:
    result = pickle.dump(npy, fp)
```

# 目标网络网络结构

```python
# 查看paddle写的网络的结构
from CGNet_paddle.paddlevision.models.CGNet import cgnet as paddle_cgnet
model2 = paddle_cgnet(pretrained=False,
                  classes=19,
                  m=3,
                  n=21)
model2
```

```python
import pickle
with open('weight.pkl', 'rb') as file:
    npy = pickle.load(file)
```

```python
len(npy), len(model2.state_dict())
```

```python
# 检查两个dict的权重格式是否一样的
i = 0
for key, value in model2.state_dict().items():
    print(value.numpy().shape, ' '*(30 - len(str(value.numpy().shape))), npy[i].shape)
    i+=1
```

```python
# 查看那些连 形状都不一样的权重出在哪个地方
i = 0
for key, value in model2.state_dict().items():
    if(value.numpy().shape != npy[i].shape):
        print(value.numpy().shape, ' '*(30 - len(str(value.numpy().shape))), npy[i].shape)
        print(key)
    i+=1
```
